# PREPROCESSING

- The World steel Association datasets were extracted from the "STEEL STATISTICAL YEARBOOK 2019.pdf", provided by World Steel Association. Thus these data have a lot of unnamed columns,null-values, unwanted spaces, unwanted charecters, etc thus we need to remove those.
- All the numeric values inside the columns have white space present inbetween them as present in the pdf, thus we need to remove all those.
- Then we have to convert Dtype: object into int or float type.
- At last we will save all those datasets for further use.

```python
import pandas as pd
import numpy as np
```

### Removing unwanted Characters

```python
# Dataset 1:
df1.head(50)
```

- Some country names have (e) at the end and Ranking columns have N/R to null values

```python
df1.columns
```

```python
# Replacing N/R to NaN and changing Dtype of column 2019 Ranking into float. 
df1['2019 Ranking'] = pd.to_numeric(df1['2019 Ranking'], errors='coerce')
```

```python
# Removing the parenthesis and anything inside parenthesis by using regex.
df1['Companies'] = df1['Companies'].str.replace(r"\(.*?\)","").copy()
df1['Country'] = df1['Country'].str.replace(r"\(.*?\)","").copy()
```

```python
# Dataset 2:
df2.tail(50)
```

- Country names have extra charecters inside parenthesis and the 2018 columns have 'e' at the end.

```python
# Removing the parenthesis by using regex.
df2['Country'] = df2['Country'].str.replace(r"\(.*?\)","").copy()
```

```python
# Removing e from the column 2018.
df2['2018'] = df2['2018'].str.replace('''e''','').copy()
```

```python
# Dataset 3:
df3.head(50)
```

- Country name has (e) which we have to remove.

```python
df3['Country'] = df3['Country'].str.replace(r"\(.*?\)",'')
```

```python
# Dataset 4:
df4.tail(50)
```

- Dataset 4 also has the same problem: country names has (e). 
- No other problems.

```python
df4['Country'] = df4['Country'].str.replace(r"\(.*?\)",'').copy()
```

```python
# Dataset 5:
df5.tail(50)
```

```python
df5['Country'] = df5['Country'].str.replace(r"\(.*?\)",'').copy()
```

```python
# Dataset 6:
df6.head(50)
```

```python
# Cleaning (1),(2),(e) and e from colunm - Country, 2018.
df6['Country'] = df6['Country'].str.replace(r"\(.*?\)",'').copy()
df6['2018'] = df6['2018'].str.replace("e",'').copy()
```

### Removing white space.

```python
#Dataset 1:
df1.head()
```

```python
# Strip whitespace at the start and the end if present
cols = ['Companies', 'Country', 'Unit']
df1[cols] = df1[cols].apply(lambda x: x.str.strip())
```

```python
# Dataset 2:
df2.head()
```

```python
df2.columns
```

```python
# Strip whitespace at the start and the end if present
cols = ['Country','Product Type', 'Division', 'Unit']
df2[cols] = df2[cols].apply(lambda x: x.str.strip())
```

```python
# Strip the white space between the numeric values.
cols = ['2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018']
df2[cols] = df2[cols].apply(lambda x: x.str.replace(' ', ''))
```

```python
# Dataset 3:
df3.head()
```

```python
df3.columns
```

```python
# Strip whitespace at the start and the end if present
cols = ['Country','Division']
df3[cols] = df3[cols].apply(lambda x: x.str.strip())
```

```python
# Strip the white space between the numeric values.
cols = [ 'OBC thousand tonnes', 'EF thousand tonnes','OHF thousand tonnes', 'Other thousand tonnes', 'Total thousand tonnes']
df3[cols] = df3[cols].apply(lambda x: x.str.replace(' ', ''))
```

```python
# Dataset 4:
df4.head(50)
```

```python
df4.columns
```

```python
# Strip whitespace at the start and the end if present
cols = ['Country','Division', 'Import/Export', 'Product Type']
df4[cols] = df4[cols].apply(lambda x: x.str.strip())
```

```python
# Strip the white space between the numeric values.
cols = [ '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018']
df4[cols] = df4[cols].apply(lambda x: x.str.replace(' ', ''))
```

```python
# Dataset 5:
df5.head()
```

```python
df5.columns
```

```python
# Strip whitespace at the start and the end if present
cols = ['Country','Division', 'Unit', 'Type']
df5[cols] = df5[cols].apply(lambda x: x.str.strip())
```

```python
# Strip the white space between the numeric values.
cols = [ '2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018']
df5[cols] = df5[cols].apply(lambda x: x.str.replace(' ', ''))
```

```python
# Dataset 6:
df6.head()
```

```python
df6.columns
```

### Changing the dtype.

```python
# Dataset 1:
df1.info()
```

```python
# Dataset 2:
df2.info()
```

```python
# Column 2009,2010...2017,2018 should be numeric value.
cols = ['2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']
df2[cols] = df2[cols].apply(pd.to_numeric)
```

```python
# Dataset 3:
df3.info()
```

```python
df3.columns
```

```python
# Some Column should be numeric value.
cols = ['OBC thousand tonnes', 'EF thousand tonnes',
       'OHF thousand tonnes', 'Other thousand tonnes', 'Total thousand tonnes', 'Year']
df3[cols] = df3[cols].apply(pd.to_numeric)
```

```python
# Changing the dtype of "Year" from float to int.
df3['Year'] = df3['Year'].astype(np.int64).copy()
```

```python
# Dataset 4:
df4.info()
```

```python
df4.columns
```

```python
# Column 2009,2010...2017,2018 should be numeric value.
# These columns have some non-null but non-numeric values like "...", which has no significance so we will remove them.
cols = ['2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018']
df4[cols] = df4[cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Non-numeric values are changed to null so we will change them to 0.
df4.isna().sum()
```

```python
df4 = df4.fillna(0)
```

```python
# Dataset 5:
df5.info()
```

```python
# Column 2009,2010...2017,2018 should be numeric value.

cols = ['2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018']
df5[cols] = df5[cols].apply(pd.to_numeric)
```

```python
# Dataset 6:
df6.info()
```

```python
# Column 2009,2010...2017,2018 should be numeric value.

cols = ['2009', '2010', '2011', '2012', '2013', '2014', '2015',
       '2016', '2017', '2018']
df6[cols] = df6[cols].apply(pd.to_numeric, errors = 'coerce')
```

```python
df6.isna().sum()
```

```python
# Replacing newly formed null values with 0. 
df6 = df6.fillna(0).copy()
```
